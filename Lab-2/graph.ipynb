```python
import networkx as nx
DG = nx.DiGraph()
DG.add_node("v1")
DG.add_node("v2")
DG.add_node("v3")
DG.add_node("v4")
DG.add_node("v5")
DG.add_node("v6")

DG.add_edge("v1", "v2")
DG.add_edge("v1", "v4")

DG.add_edge("v2", "v5")

DG.add_edge("v3", "v1")
DG.add_edge("v3", "v4")

DG.add_edge("v5", "v3")

DG.add_edge("v6", "v1")
DG.add_edge("v6", "v3")
DG.add_edge("v6", "v4")
```

```python
import matplotlib.pyplot as plt
```

```python
nx.draw(DG, with_labels=True, font_weight='bold', node_size=500, node_color='lightblue', font_size=15, arrowsize=20)
```

```python
adj_matrix = nx.adjacency_matrix(DG)
print("邻接矩阵：")
print(adj_matrix.todense())
incidence_matrix = nx.incidence_matrix(DG)
print("关联矩阵：")
print(incidence_matrix.todense())

# 边列表
edge_list = list(DG.edges())
print("边列表：")
print(edge_list)

# 正向邻接表
adj_list = nx.to_dict_of_lists(DG)
print("正向邻接表：")
print(adj_list)

# 邻接字典
adj_dict = nx.to_dict_of_dicts(DG)
print("邻接字典：")
print(adj_dict)
```
