```python
import networkx as nx
DG = nx.DiGraph()
DG.add_node("v1")
DG.add_node("v2")
DG.add_node("v3")
DG.add_node("v4")
DG.add_node("v5")
DG.add_node("v6")

DG.add_edge("v1", "v2")
DG.add_edge("v1", "v4")

DG.add_edge("v2", "v5")

DG.add_edge("v3", "v1")
DG.add_edge("v3", "v4")

DG.add_edge("v5", "v3")

DG.add_edge("v6", "v1")
DG.add_edge("v6", "v3")
DG.add_edge("v6", "v4")
```

```python
import matplotlib.pyplot as plt
```

```python
nx.draw(DG, with_labels=True, font_weight='bold', node_size=500, node_color='lightblue', font_size=15, arrowsize=20)
```

```python

DG_2 = nx.DiGraph()
# node a, b, c, d, e, f
DG_2.add_node("a")
DG_2.add_node("b")
DG_2.add_node("c")
DG_2.add_node("d")
DG_2.add_node("e")
DG_2.add_node("f")
DG_2.add_edge("b", "a")
DG_2.add_edge("c", "b")
DG_2.add_edge("d", "c")
DG_2.add_edge("f", "b")
DG_2.add_edge("f", "c")
DG_2.add_edge("f", "e")
DG_2.add_edge("b", "e")
DG_2.add_edge("c", "e")
DG_2.add_edge("a", "d")
nx.draw(DG_2, with_labels=True, font_weight='bold', node_size=500, node_color='lightblue', font_size=15, arrowsize=20)
plt.show()
isomorphic = nx.is_isomorphic(DG, DG_2)
print("", isomorphic)
GM = nx.isomorphism.GraphMatcher(DG, DG_2)
match = GM.mapping
print("", match)
```

```python
import networkx as nx

# 
G1 = nx.Graph()
G1.add_edges_from([(1, 2), (2, 3), (3, 4)])

G2 = nx.Graph()
G2.add_edges_from([('a', 'b'), ('b', 'c'), ('c', 'd')])

#  isomorphism 
GM = nx.isomorphism.GraphMatcher(G1, G2)

# 
if GM.is_isomorphic():
    print(" G1  G2 ")
    # 
    match = GM.mapping
    print(":", match)
else:
    print(" G1  G2 ")
```
