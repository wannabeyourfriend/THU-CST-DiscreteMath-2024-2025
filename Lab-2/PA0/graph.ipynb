```python
import networkx as nx
DG = nx.DiGraph()
DG.add_node("v1")
DG.add_node("v2")
DG.add_node("v3")
DG.add_node("v4")
DG.add_node("v5")
DG.add_node("v6")

DG.add_edge("v1", "v2")
DG.add_edge("v1", "v4")

DG.add_edge("v2", "v5")

DG.add_edge("v3", "v1")
DG.add_edge("v3", "v4")

DG.add_edge("v5", "v3")

DG.add_edge("v6", "v1")
DG.add_edge("v6", "v3")
DG.add_edge("v6", "v4")
```

```python
import matplotlib.pyplot as plt
```

```python
nx.draw(DG, with_labels=True, font_weight='bold', node_size=500, node_color='lightblue', font_size=15, arrowsize=20)
```

```python

DG_2 = nx.DiGraph()
# node： a, b, c, d, e, f
DG_2.add_node("a")
DG_2.add_node("b")
DG_2.add_node("c")
DG_2.add_node("d")
DG_2.add_node("e")
DG_2.add_node("f")
DG_2.add_edge("b", "a")
DG_2.add_edge("c", "b")
DG_2.add_edge("d", "c")
DG_2.add_edge("f", "b")
DG_2.add_edge("f", "c")
DG_2.add_edge("f", "e")
DG_2.add_edge("b", "e")
DG_2.add_edge("c", "e")
DG_2.add_edge("a", "d")
nx.draw(DG_2, with_labels=True, font_weight='bold', node_size=500, node_color='lightblue', font_size=15, arrowsize=20)
plt.show()
isomorphic = nx.is_isomorphic(DG, DG_2)
print("两个图是否同构：", isomorphic)
GM = nx.isomorphism.GraphMatcher(DG, DG_2)
match = GM.mapping
print("同构映射：", match)
```

```python
import networkx as nx

# 创建两个图
G1 = nx.Graph()
G1.add_edges_from([(1, 2), (2, 3), (3, 4)])

G2 = nx.Graph()
G2.add_edges_from([('a', 'b'), ('b', 'c'), ('c', 'd')])

# 使用 isomorphism 模块检查同构性
GM = nx.isomorphism.GraphMatcher(G1, G2)

# 检查是否同构
if GM.is_isomorphic():
    print("图 G1 和 G2 是同构的")
    # 返回匹配
    match = GM.mapping
    print("匹配关系:", match)
else:
    print("图 G1 和 G2 不是同构的")
```

```python
adj_matrix = nx.adjacency_matrix(DG)
print("邻接矩阵：")
print(adj_matrix.todense())
incidence_matrix = nx.incidence_matrix(DG)
print("关联矩阵：")
print(incidence_matrix.todense())

# 边列表
edge_list = list(DG.edges())
print("边列表：")
print(edge_list)

# 正向邻接表
adj_list = nx.to_dict_of_lists(DG)
print("正向邻接表：")
print(adj_list)

# 邻接字典
adj_dict = nx.to_dict_of_dicts(DG)
print("邻接字典：")
print(adj_dict)
```
